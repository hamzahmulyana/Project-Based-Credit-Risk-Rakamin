<h4 style='text-align:left; color:cyan'><b> Success Criteria </b></h4>

1. Miss Prediction

    Mengingat fokus pada proyek ini adalah untuk mengurangi orang yang tidak bisa/gagal membayar, maka akan ditetapkan sebagai berikut:
    Target:
    - 0: Ditolak
    - 1: Disetujui

    **Kesalahan Prediksi:**

    Dalam konteks pengembangan machine learning untuk memprediksi risiko credit, evaluasi metrik memainkan peran penting dalam menilai kinerja model dan dampak potensianya terhadap pendapatan perusahaan:
    
    |                    | **Ditolak (0)**            | **Disetujui (1)**             |
    |--------------------|-------------------------|-----------------------|
    | **Ditolak (0)**   | **True Negative (TN)**  | **False Positive (FP)**   |
    |                    | Prediction: Ditolak    | Prediction: Disetujui     |
    |                    | Actual: Ditolak      | Actual: Ditolak      |
    | **Disetujui (1)**      | **False Negative (FN)**     | **True Positive (TP)**    |
    |                    | Prediction: Ditolak    | Prediction: Disetujui     |
    |                    | Actual: Disetujui          | Actual: Disetujui         |

    **Type 1 Error (False Positive):** Model tersebut memprediksi bahwa orang tersebut layak mendapatkan pinjaman, akan tetapi kenyataannya orang tersebut tidak layak mendapatkan pinjaman.

    **Contoh kasus:**

    Bayangkan seorang nasabah bernama David mengajukan pinjaman sebesar $10.000. Berdasarkan data yang ada, model memprediksi David sebagai "layak" sehingga bank menyetujui pinjamannya. Namun, ternyata David tidak pernah melakukan penyetoran cicilan sama sekali.
    
    **Akibat kesalahan prediksi ini:**
    - Biaya Pemasaran yang Terbuang
        - Bank telah melakukan 3 kali panggilan telepon untuk menawarkan produk lain, menghabiskan waktu dan sumber daya staf.
        - Bank juga mengirim 3 email promosi kepada David.

    - Kerugian Pokok Pinjaman
        - Karena David tidak membayar sama sekali, bank kehilangan seluruh pokok pinjaman sebesar $10,000.

    - Bunga yang Tidak Tertagih
        - Jika seharusnya bunga pinjaman adalah 12% per tahun, bank kehilangan potensi pendapatan bunga sebesar $1,200.

    **Total Kerugian:**
    - Pokok pinjaman: $10,000
    - Bunga tidak tertagih: $1,200
    - Total: $11,200


    **Type 2 Error (False Negative):** Model tersebut memprediksi bahwa orang tersebut tidak layak mendapaykan pinjaman, akan tetapi kenyataannya orang tersebut layak mendapatkan pinjaman.

    **Contoh kasus:**

    Lisa, seorang wiraswasta dengan penghasilan stabil  $5,000/bulan, mengajukan pinjaman $10,000. Namun, model menolaknya karena tidak memiliki slip gaji tetap. Padahal, Lisa sebenarnya mampu membayar tepat waktu.

    **Akibat kesalahan prediksi ini:**
    - Kehilangan Pendapatan Bunga
        - Bank kehilangan potensi bunga 12% per tahun ($1,200).
    - Biaya Evaluasi yang Sia-sia
        - Tim analis telah menghabiskan waktu memeriksa dokumen Lisa.
    - Kehilangan Pelanggan Potensial
        - Lisa mungkin beralih ke kompetitor dan tidak kembali ke bank tersebut.

    **Total Kerugian:**
    - Bunga tidak didapat: $1,200
    - Total: $1,2000

```python
def calculate_adjacent_date(date, operation='next'):
    """Fungsi universal untuk menghitung tanggal sebelum/sesudah"""
    # Daftar bulan untuk referensi
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

    if pd.isna(date):
        return None
    
    try:
        month, day = date.split('-')
        current_idx = months.index(month)
        
        if operation == 'next':
            new_idx = (current_idx + 1) % 12
        elif operation == 'prev':
            new_idx = (current_idx - 1) % 12
        else:
            return None
            
        return f"{months[new_idx]}-{day}"
    except:
        return None

# Pertama isi next_pymnt_d yang kosong dimana last_pymnt_d ada
next_mask = df['next_pymnt_d'].isna() & df['last_pymnt_d'].notna()
df.loc[next_mask, 'next_pymnt_d'] = df.loc[next_mask, 'last_pymnt_d'].apply(lambda x: calculate_adjacent_date(x, 'next'))

# Kemudian isi last_pymnt_d yang kosong dimana next_pymnt_d ada
prev_mask = df['last_pymnt_d'].isna() & df['next_pymnt_d'].notna()
df.loc[prev_mask, 'last_pymnt_d'] = df.loc[prev_mask, 'next_pymnt_d'].apply(lambda x: calculate_adjacent_date(x, 'prev'))
```

```python
# Fungsi untuk analisis fitur
def analyze_features(df, target='loan_status', alpha=0.05):
    results = []
    
    numeric_col_new_df = numeric_col_new.drop(columns=[target], errors='ignore')
    # Uji fitur numerik (Mann-Whitney U)
    for feature in numeric_col_new_df.columns:
        group1 = df[feature][df[target] == 1]
        group0 = df[feature][df[target] == 0]
        
        if len(group1) > 0 and len(group0) > 0:
            stat, p = mannwhitneyu(group1, group0, alternative='two-sided')
            is_selected = 'Ya' if p < alpha else 'Tidak'
            results.append({
                'Fitur': feature,
                'Jenis Data': 'Numerik',
                'Uji Statistik': 'Mann-Whitney U',
                'P-value': f"{p:.4f}",
                'Signifikan': 'Tolak H0' if p < alpha else 'Terima H0',
                'Dipilih': is_selected
            })
    
    # Uji fitur kategorik (Chi-Square)
    categorical_features_new_df = categorical_col_new
    for feature in categorical_features_new_df.columns:
        contingency_table = pd.crosstab(df[feature], df[target])
        if contingency_table.size > 0:
            try:
                chi2, p, dof, expected = chi2_contingency(contingency_table)
                is_selected = 'Ya' if p < alpha else 'Tidak'
                results.append({
                    'Fitur': feature,
                    'Jenis Data': 'Kategorik',
                    'Uji Statistik': 'Chi-Square',
                    'P-value': f"{p:.4f}",
                    'Signifikan': 'Tolak H0' if p < alpha else 'Terima H0',
                    'Dipilih': is_selected
                })
            except ValueError:
                pass  # Skip jika ada nilai yang tidak valid
    
    return pd.DataFrame(results)
```

```python
# Build class for data Feature Creation process
class FeatureCreation(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        # Feature creation
        X['mnth_issue_d'] = X['issue_d'].str.slice(0, 3)
        X['day_issue_d'] = X['issue_d'].str[-2:].astype(int)

        X['mnth_last_pymnt_d'] = X['last_pymnt_d'].str.slice(0, 3)
        X['day_last_pymnt_d'] = X['last_pymnt_d'].str[-2:].astype(int)

        X['mnth_next_pymnt_d'] = X['next_pymnt_d'].str.slice(0, 3)
        X['day_next_pymnt_d'] = X['next_pymnt_d'].str[-2:].astype(int)

        return X

# Build class for Detection Outliers
class OutlierDetection(BaseEstimator, TransformerMixin):
    def __init__(self, lower_percentile=10, upper_percentile=90):
        self.lower_percentile = lower_percentile
        self.upper_percentile = upper_percentile
        self.lower_bound = None
        self.upper_bound = None

    def fit(self, X, y=None):
        X = np.asarray(X)
        self.lower_bound = np.percentile(X, self.lower_percentile)
        self.upper_bound = np.percentile(X, self.upper_percentile)
        return self

    def transform(self, X):
        X = np.asarray(X)
        X = np.where(X < self.lower_bound, self.lower_bound, X)
        X = np.where(X > self.upper_bound, self.upper_bound, X)
        return X
    
# Build class for Selector Columns
class ColumnSelector(BaseEstimator, TransformerMixin):
    def __init__(self, columns=None):
        self.columns = columns
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X
    
# Ordinal mapping
ordinal_mapping_select_feature = [
    {'col':'term', 'mapping':{'36 months':1, '60 months':2, None:0}},
    {'col':'emp_length', 'mapping':{'< 1 year':1, '1 year':2, '2 years':3, '3 years':4, '4 years':5, '5 years':6, '6 years':7, '7 years':8, '8 years':9, '9 years':10, '10+ years':11, None:0}},
    {'col':'grade', 'mapping':{'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7, None:0}},
    {'col':'mnth_issue_d', 'mapping':{'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12, None:0}},
    {'col':'mnth_last_pymnt_d', 'mapping':{'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12, None:0}},
    {'col':'mnth_next_pymnt_d', 'mapping':{'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12, None:0}},
]

ordinal_mapping = [
    {'col':'mnth_last_pymnt_d', 'mapping':{'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12, None:0}},
    {'col':'mnth_next_pymnt_d', 'mapping':{'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12, None:0}},
]

# Pipeline Feature Engineering
robust_preprocessor_outlier = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('outlier', OutlierDetection()),
    ('scaler', RobustScaler())])

robust_preprocessor = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler())])

onehot_preprocessor = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'))])

ordinal_preprocessor_select_feature = Pipeline(steps=[
    ('ordinal', OrdinalEncoder(mapping=ordinal_mapping_select_feature))])

ordinal_preprocessor = Pipeline(steps=[
    ('ordinal', OrdinalEncoder(mapping=ordinal_mapping))])

feature_preprocessor = Pipeline(steps=[
    ('selectfeature', ColumnSelector())])


# Transformer from Pipeline Feature Engineering
preprocessor1 = ColumnTransformer(
    transformers=[
        ("onehot", onehot_preprocessor, 
         ['home_ownership', 'verification_status', 'purpose', 'addr_state', 'initial_list_status']),
        ("ordinal", ordinal_preprocessor_select_feature, 
         ['term', 'emp_length', 'grade', 'mnth_issue_d', 'mnth_last_pymnt_d', 'mnth_next_pymnt_d']),
        ("robust", robust_preprocessor, 
         ['loan_amnt', 'dti', 'revol_util']),
        ("robust_outlier", robust_preprocessor_outlier, 
         ['total_rec_int', 'open_acc', 'pub_rec', 'int_rate', 'installment', 'annual_inc', 'revol_bal', 
          'total_acc', 'out_prncp', 'total_pymnt', 'total_rec_prncp', 'total_rec_late_fee', 'recoveries', 
          'collection_recovery_fee', 'last_pymnt_amnt', 'tot_cur_bal']),
        ("colomselect",feature_preprocessor, 
         ['delinq_2yrs', 'inq_last_6mths', 'day_issue_d', 'day_last_pymnt_d', 'day_next_pymnt_d'])
    ],
    remainder='drop', verbose_feature_names_out=False
)
```

<h2 style='text-align:left; color:darkblue'><b> Modeling Experiment </b></h1>

Pada Experiment kali ini Metric Evaluation yang digunakan adalah Precision, karena pada experiment kali ini akan menekankan kesalahan Type 1 Error.
- **Precision**
    
    **Precision** adalah metrik evaluasi yang digunakan untuk mengukur akurasi prediksi kelas positif dalam model klasifikasi. Precision menjawab pertanyaan:
    "Dari semua prediksi positif yang dibuat model, berapa persen yang benar-benar positif?"
    
    Rumus Menghitung Precision:
    $$
    \text{Precision} = \frac{TP}{TP + FP}
    $$

    di mana:
    - **TP (True Positive)**: Prediksi benar sebagai kelas positif.
    - **FP (False Positif)**: Prediksi salah sebagai kelas positif (observasi sebenarnya negatif). 

    **Penjelasan Rumus Precison**

    1. Identifikasi Prediksi Positif:

        Hitung semua prediksi model yang termasuk kelas positif (TP + FP).

    2. Hitung Rasio Prediksi Benar:

        Bagi jumlah prediksi benar positif (TP) dengan total prediksi positif (TP + FP).

    3. Interpretasi:
        - Precision = 1.0 (100%): Model tidak pernah salah memprediksi positif (tidak ada FP).
        - Precision = 0.0: Semua prediksi positif model salah.

**Create Function for Evaluation Model with Cross Validation**

```python
def evaluate_models_with_resampling(models, X_train, y_train, scoring, 
                                  resampling_methods=None, 
                                  no_resampling=False):
    """
    Evaluasi model dengan berbagai metode resampling
    
    Parameters:
    -----------
    models : dict
        Dictionary berisi model-model yang akan dievaluasi (key: model object, value: nama model)
    X_train : DataFrame
        Data training features
    y_train : Series
        Data training target
    scoring : dict
        Metric scoring untuk evaluasi
    resampling_methods : dict, optional
        Dictionary berisi metode resampling (key: sampler object, value: nama sampler)
        Default: None (tidak menggunakan resampling)
    no_resampling : bool, optional
        Jika True, akan menambahkan evaluasi tanpa resampling sebagai baseline
        Default: False
    """
    
    results = {
        'Model': [],
        'Resampling': [],
        'mean_precision_train': [],
        'std_precision_train': [],
        'mean_precision_test': [],
        'std_precision_test': [],
    }
    
    # Jika no_resampling=True, tambahkan evaluasi tanpa resampling
    if no_resampling:
        resampling_methods = resampling_methods or {}
        resampling_methods[None] = 'No Resampling'
    
    # Jika tidak ada resampling methods yang diberikan
    if resampling_methods is None:
        resampling_methods = {None: 'No Resampling'}
    
    for model, model_name in models.items():
        for sampler, sampler_name in resampling_methods.items():
            # Buat pipeline
            steps = [
                ('feature_creation', FeatureCreation()),
                ('preprocessor', preprocessor),
            ]
            
            # Tambahkan resampling jika ada
            if sampler is not None:
                steps.append(('resampling', sampler))
            
            steps.append(('model', model))
            
            pipeline = ImbPipeline(steps=steps)
            
            # Cross-validation
            cv_score = cross_validate(
                estimator=pipeline,
                X=X_train,
                y=y_train,
                scoring=scoring,
                cv=5,
                return_train_score=True,
                error_score='raise'
            )
            
            # Simpan hasil
            results['Model'].append(model_name)
            results['Resampling'].append(sampler_name)
            results['mean_precision_train'].append(np.mean(cv_score['train_precision']))
            results['std_precision_train'].append(np.std(cv_score['train_precision']))
            results['mean_precision_test'].append(np.mean(cv_score['test_precision']))
            results['std_precision_test'].append(np.std(cv_score['test_precision']))
    
    return pd.DataFrame(results)
```

**Create Function to Randomized Search CV**

```python
def randomized_search(pipeline, parameters, scoring, X_train, y_train):
    # Menentukan RandomizedSearchCV dengan pipeline dan parameter distribusi
    randomized_search = RandomizedSearchCV(
        estimator=pipeline,
        param_distributions=parameters,
        n_iter=5,
        refit='precision',
        scoring=scoring,
        cv=5,
        random_state=42,
        return_train_score=True
    )
    
    # Melatih RandomizedSearchCV dengan data pelatihan
    randomized_search.fit(X_train, y_train)
    
    return randomized_search
```

**Create Function to View Result of cv_result_**

```python
# Create function
def cv_result(cv_results, metrics):
    # Konversi hasil cv_results ke DataFrame
    cv_results_df = pd.DataFrame(cv_results)
    
    results = {}
    
    for metric in metrics:
        train_mean_col = f'mean_train_{metric}'
        train_std_col = f'std_train_{metric}'
        test_mean_col = f'mean_test_{metric}'
        test_std_col = f'std_test_{metric}'
        
        # Mean and std for metrik train
        results[f'mean_{metric}_train'] = [cv_results_df[train_mean_col].mean()] if metric in ['precision'] else [cv_results_df[train_mean_col].mean()]
        results[f'std_{metric}_train'] = [cv_results_df[train_std_col].mean()]
        
        # Mean and std for metrik test
        results[f'mean_{metric}_test'] = [cv_results_df[test_mean_col].mean()] if metric in ['precision'] else [cv_results_df[test_mean_col].mean()]
        results[f'std_{metric}_test'] = [cv_results_df[test_std_col].mean()]

    return pd.DataFrame(results)
```
